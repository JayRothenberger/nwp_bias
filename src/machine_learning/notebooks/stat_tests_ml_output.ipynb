```python
df_out = df_out.sort_index()
```

```python
df_out["valid_time"] = valid_time[:-5]
df_out["t2m_OLEA"] = temp[:-5]
df_out["tcc_OLEA"] = tcc[:-5]
df_out["u"] = u[:-5]
df_out["slp"] = slp[:-5]
df_out["tp"] = precip[:-5]
df_out["snow"] = snow[:-5]
```

```python
def groupby_month(df, col):
    df = df[df[col] > -999]
    GB = df.groupby([(df.valid_time.dt.month)])[col].mean()
    the_list = GB.tolist()
    fig, ax = plt.subplots(figsize=(21, 6))
    x = np.arange(1, 13)
    plt.bar(x, the_list)
    plt.ylim(-0.2, 0.2)
    # ax.set_xticklabels([2018, 2019, 2020, 2021, 2022])
    # ax.set_xticks(np.arange(0, 60, 12))
    return the_list
```

```python
# Developing a function to round to a multiple
def round_to_multiple(number, multiple):
    return multiple * round(number / multiple)
```

```python
df_out["t2m_OLEA"] = round_to_multiple(df_out["t2m_OLEA"], 5)
df_out["tcc_OLEA"] = round_to_multiple(df_out["tcc_OLEA"], 5)
df_out["slp"] = round_to_multiple(df_out["slp"], 10)
df_out["u"] = round_to_multiple(df_out["u"], 1)
df_out["snow"] = df_out["snow"].round(3)
# df_out['snow'] = round_to_multiple(df_out['snow'], 0.05)
```
