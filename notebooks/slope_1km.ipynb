```python
path = "/home/aevans/landtype/elevation/data/NY/slope/1km"
directory = sorted(os.listdir(path))
directory
```

```python
mode_ls = []
for i, _ in enumerate(directory):
    df = pd.read_csv(f"{path}/{directory[i]}")
    mode = df["VALUE"].mode()
    if len(mode) > 1:
        mode = mode[1]
    else:
        mode = mode[0]
    mode_ls.append(mode)
```

```python
mode_ls
```

```python
df_1km = pd.DataFrame()
df_1km["station"] = station_list_ny
df_1km["aspect"] = mode_ls
df_1km["lat"] = ny_df_lats
df_1km["lon"] = ny_df_lons
df_1km[110:126]
```
